````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Fixing random state for reproducibility
np.random.seed(2360873)
%matplotlib inline
````

````python
df = pd.read_csv('previsao_de_renda.csv')

df.drop(['Unnamed: 0'], axis=1, inplace=True)

df
````

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
# Extrair o ano da coluna 'data_ref' usando slicing
df['ano'] = df['data_ref'].str[:4]

df['posse_de_veiculo'] = df['posse_de_veiculo'].astype(str)
df['posse_de_imovel'] = df['posse_de_imovel'].astype(str)
df['mau'] = df['mau'].astype(str)

# Selecionar as colunas qualitativas para análise
colunas_qualitativas = ['sexo', 'posse_de_veiculo', 'posse_de_imovel', 'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia', 'mau']

# Criar um gráfico de barras para cada variável qualitativa ao longo do tempo
plt.figure(figsize=(20, 12))
for coluna in colunas_qualitativas:
    plt.subplot(3, 3, colunas_qualitativas.index(coluna) + 1)
    sns.countplot(data=df, x='ano', hue=coluna, palette='Set2')
    plt.title(f'{coluna} ao Longo do Tempo')
    plt.xlabel('Ano')
    plt.ylabel('Contagem')

plt.tight_layout()
plt.show()

````

A obtenção de clientes teve uma queda significativa de um ano para outro, resultando em uma queda expressiva em todas as outras variáveis qualitativas também.

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
# Criar gráfico de ponto para cada variável qualitativa mostrando as médias por ano
# Calcular as médias de cada nível das variáveis qualitativas por ano
df['posse_de_veiculo'] = df['posse_de_veiculo'].astype(str)
df['posse_de_imovel'] = df['posse_de_imovel'].astype(str)
df['mau'] = df['mau'].astype(str)

mean_df = df.groupby(['ano', 'sexo', 'posse_de_veiculo', 'posse_de_imovel', 'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia', 'mau']).size().unstack().mean(axis=1)

# Converter o resultado em DataFrame
mean_df = mean_df.reset_index(name='media')

plt.figure(figsize=(12, 8))
for coluna in colunas_qualitativas:
    ax = sns.pointplot(x="ano", y="media", hue=coluna,
                       data=mean_df, dodge=True, ci=95)
    plt.title(f'Média de {coluna} por Ano')
    plt.xlabel('Ano')
    plt.ylabel('Média')
    plt.legend(title=coluna)
    plt.show()
````

Observa-se que a queda das medias das variavéis qualitativas mantiveram uma proporção satisfatória, o que representa uma boa estabilidade ao longo do tempo. Portando essas variáveis podem ser utilizadas em uma possível modelagem.
