#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('sinasc_RO_2019.csv', na_values=('NaN', 'na', ' ' ))
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python

fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Gráfico de barras para APGAR1
axs[0, 0].bar(df['APGAR1'].value_counts().index, df['APGAR1'].value_counts().values, color='blue')
axs[0, 0].set_title('APGAR1')

# Gráfico de barras para APGAR5
axs[0, 1].bar(df['APGAR5'].value_counts().index, df['APGAR5'].value_counts().values, color='red')
axs[0, 1].set_title('APGAR5')

# Gráfico de barras para SEXO
axs[0, 2].bar(df['SEXO'].value_counts().index, df['SEXO'].value_counts().values, color='green')
axs[0, 2].set_title('SEXO')

# Gráfico de barras para GESTACAO
axs[1, 0].bar(df['GESTACAO'].value_counts().index, df['GESTACAO'].value_counts().values, color='purple')
axs[1, 0].set_title('GESTACAO')
axs[1, 0].tick_params(axis='x', labelrotation=90)

# Gráfico de barras para ESTCIVMAE
axs[1, 1].bar(df['ESTCIVMAE'].value_counts().index, df['ESTCIVMAE'].value_counts().values, color='orange')
axs[1, 1].set_title('ESTCIVMAE')

# Ajustes finais de layout
plt.tight_layout()
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
apg5_medio_por_consultas = df.groupby('CONSULTAS')['APGAR5'].mean()

# Plotar o gráfico de barras
plt.figure(figsize=(10, 6))
apg5_medio_por_consultas.plot(kind='bar')
plt.xlabel('Quantidade de Consultas')
plt.ylabel('Média do APGAR5')
plt.yticks(range(0, 11, 1)) 
plt.title('Média do APGAR5 por Quantidade de Consultas')
plt.xticks(rotation=0, ha='right') 
plt.tight_layout()
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plotar o gráfico de frequência para SEXO
df['SEXO'].value_counts().plot(kind='bar', ax=axs[0, 0])
axs[0, 0].set_title('Frequência de SEXO')

# Plotar o gráfico de frequência para GESTACAO
df['GESTACAO'].value_counts().plot(kind='bar', ax=axs[0, 1])
axs[0, 1].set_title('Frequência de GESTACAO')

# Plotar o gráfico de frequência para ESTCIVMAE
df['ESTCIVMAE'].value_counts().plot(kind='bar', ax=axs[1, 0])
axs[1, 0].set_title('Frequência de ESTCIVMAE')

# Plotar o gráfico de frequência para CONSULTAS
df['CONSULTAS'].value_counts().plot(kind='bar', ax=axs[1, 1])
axs[1, 1].set_title('Frequência de CONSULTAS')

plt.tight_layout()
plt.show()
```
