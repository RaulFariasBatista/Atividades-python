## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos_Matemática'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos_Português'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos_Geografia'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos_Inglês'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos_História'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos_Física'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos_Química'])
````

````python

tabela_acertos = pd.concat([df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui], axis=1)

tabela_acertos
````

````python
# 1) Seu código aqui
pc_mat = df_mat / 24
pc_por = df_por / 18
pc_geo = df_geo /  8
pc_ing = df_ing /  8
pc_his = df_his /  8
pc_fis = df_fis /  8
pc_qui = df_qui /  8
tabela_porcentagem = round(pd.concat([pc_mat, pc_por, pc_geo, pc_ing, pc_his, pc_fis, pc_qui], axis=1),3)
tabela_porcentagem.columns = ['%_acertos_Matemática', '%_acertos_Português', '%_acertos_Geografia',
                              '%_acertos_Inglês', '%_acertos_História', '%_acertos_Física', '%_acertos_Química']
tabela_porcentagem
````

````python
# 2) Seu código aqui

tabela_acertos['Total_acertos'] = tabela_acertos.iloc[:, 0:].sum(axis=1)

df_total_acertos = tabela_acertos[['Total_acertos']]

df_total_acertos

````

````python
# 3) Seu código aqui

Qt_questoes = 90

porcentagem_acertos['% Acertos'] = round(df_total_acertos / Qt_questoes * 100, 3)

porcentagem_acertos
````

````python
# 4) Seu código aqui
alunos_aprovados = df_total_acertos[df_total_acertos['Total_acertos'] > 45]

num_alunos_aprovados = len(alunos_aprovados)

print(f"Número de alunos com nota maior que 45: {num_alunos_aprovados}")
````
