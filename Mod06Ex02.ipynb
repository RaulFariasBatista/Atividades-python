```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('sinasc_RO_2019.csv', na_values=('NaN', 'na', ' ' ))
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
plt.figure(figsize=(8, 6))  # Tamanho da figura
df[['IDADEMAE']].boxplot(vert=True, widths=0.7, patch_artist=True, boxprops=dict(facecolor='pink', color='black'), medianprops=dict(color='purple'))
plt.ylabel('Idade')
plt.title('Idade da mãe')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.figure(figsize=(8, 6))  # Tamanho da figura
df[['PESO']].boxplot(vert=True, widths=0.7, patch_artist=True, boxprops=dict(facecolor='lightblue', color='black'), medianprops=dict(color='red'))
plt.ylabel('Peso (g)')
plt.title('Boxplot do Peso ao Nascer')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.figure(figsize=(12, 6))
df['IDADEMAE'].plot.hist(range=(10, 55), bins=45, color='skyblue', edgecolor='black')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.title('Histograma da Idade da Mãe')

plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
plt.figure(figsize=(12, 6))
df['PESO'].plot.hist(range=(0, 6000), bins=120, color='red', edgecolor='black')
plt.xlabel('Peso do bebê')
plt.ylabel('Frequência')
plt.title('Histograma do peso do bebê')

plt.show()
```
