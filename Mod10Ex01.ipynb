Para isto, conduziu um estudo com alguns clientes, comprovando suas rendas através de olerites e outros documentos, e pretende construir um modelo preditivo para esta renda com base em algumas variáveis que já possui em seu banco de dados.

As variáveis são intuitivas - note que há uma variável 'index' que é um identificador do cliente e que em geral o ```read_csv``` do pandas coloca também uma variável sequencial.

Estes dados estão no arquivo ```previsao_de_renda.csv```. Carregue-o em um *dataframe*.

1) Avalie a estrutura de correlação das variáveis quantitativas através de um gráfico de "matriz de dispersão" e através da avaliação gráfica da matriz de correlações. Veja se você identifica algum padrão interessante ou que te faça sentido.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = pd.read_csv('previsao_de_renda.csv')
df
````

````python
numeric_columns = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_columns.corr()

# Selecionar apenas as colunas desejadas da matriz de correlação
correlation_matrix_2 = correlation_matrix[['qtd_filhos', 'idade', 'tempo_emprego', 'qt_pessoas_residencia', 'renda']]

# Remover as linhas 'Unnamed: 0' e 'index'
correlation_matrix_2.drop(['Unnamed: 0', 'index'], inplace=True)

# Exibir a matriz de correlação resultante
correlation_matrix_2
````

````python
numeric_columns.drop(['Unnamed: 0', 'index'], axis=1, inplace=True)
````

````python
sns.pairplot(numeric_columns)
plt.show()
````

````python
cmap = sns.diverging_palette(h_neg=125, h_pos=350, as_cmap=True, sep = 60, center = 'light')

sns.clustermap(numeric_columns.corr(), figsize=(10, 10), center = 0, cmap=cmap)
````

A correlação é simetrica em relação a diagonal principal. E há uma dispersão grande na maioria dos graficos, apresentando correlações baixas. 

2) Avalie um gráfico de dispersão (*scatterplot*) específico para as duas variáveis mais correlacionadas com *renda*.

````python
# Selecionar as duas variáveis mais correlacionadas com a renda
most_correlated_vars = ['renda', 'tempo_emprego' ]
scatterplot_data = df[most_correlated_vars]

# Criar o gráfico de dispersão (scatterplot)
sns.scatterplot(data=scatterplot_data, x='tempo_emprego', y='renda')
plt.title('Scatterplot de Tempo de Emprego vs. Renda')
plt.xlabel('Tempo de Emprego')
plt.ylabel('Renda')
plt.show()
````

4) Calcule o logaritmo da variável renda e repita a sequência de análise

````python
df['renda_log'] = np.log(df['renda'])

# Selecionar as colunas relevantes
numeric_columns_log = df[['qtd_filhos', 'idade', 'tempo_emprego', 'qt_pessoas_residencia', 'renda_log']]

# Calcular a matriz de correlação
correlation_matrix_log = numeric_columns_log.corr()

# Exibir a matriz de correlação resultante
correlation_matrix_log

````

````python
# Avaliar um gráfico de dispersão para as duas variáveis mais correlacionadas com "renda_log"
sns.pairplot(numeric_columns_log)
````

````python
cmap = sns.diverging_palette(h_neg=125, h_pos=350, as_cmap=True, sep = 60, center = 'light')

sns.clustermap(numeric_columns_log.corr(), figsize=(10, 10), center = 0, cmap=cmap)
````

Aplicar log apenas na variável renda não resultou em mudanças significativas. Na verdade, a correlação com o tempo de emprego, que é a mais correlacionada com a renda, apresentou uma leve queda após a aplicação do log somente na variável renda.
