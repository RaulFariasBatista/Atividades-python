````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = pd.read_csv('previsao_de_renda.csv')
df
````

````python
ax = sns.pointplot(x="posse_de_imovel", y="renda",
                   data=df, dodge=True, ci = 90)
````

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda",
                   data=df, dodge=True, ci = 90)
````

````python
df['Posse_veiculo_e_imovel'] = np.where((df['posse_de_veiculo'] == True) & (df['posse_de_imovel'] == True), True, False)
df
````

````python
ax = sns.pointplot(x="Posse_veiculo_e_imovel", y="renda",
                   data=df, dodge=True, ci = 90)
````
